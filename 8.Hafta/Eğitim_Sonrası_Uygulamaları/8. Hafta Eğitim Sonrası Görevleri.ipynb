# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(boy, kilo):
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def bilgileri_al():
    ad_soyad = input("Adınız ve soyadınız: ")
    yas = int(input("Yaşınız: "))
    boy = float(input("Boyunuz (metre cinsinden, örnek 1.75): "))
    kilo = float(input("Kilonuz (kg cinsinden): "))
    return ad_soyad, yas, boy, kilo

def kayit_yap(veri):
    with open("vki_kayitlari.txt", "a", encoding="utf-8") as dosya:
        dosya.write(veri + "\n")

def main():
    print("VKİ Hesaplama ve Kayıt Sistemine Hoş Geldiniz!")
    kayit_sayisi = int(input("Kaç kişi için giriş yapacaksınız? "))
    
    for i in range(kayit_sayisi):
        print(f"\n{i+1}. kişinin bilgileri:")
        ad_soyad, yas, boy, kilo = bilgileri_al()
        vki = vki_hesapla(boy, kilo)
        sinif = vki_siniflandir(vki)
        
        print(f"\nSonuçlar:")
        print(f"Ad-Soyad: {ad_soyad}")
        print(f"Yaş: {yas}")
        print(f"Boy: {boy} m")
        print(f"Kilo: {kilo} kg")
        print(f"VKİ: {vki:.2f}")
        print(f"Sınıf: {sinif}")
        
        kayit_verisi = f"{ad_soyad}, {yas}, {boy}m, {kilo}kg, VKİ:{vki:.2f}, {sinif}"
        kayit_yap(kayit_verisi)
        
        print("\nBilgileriniz kaydedildi!")

    print("\nTüm kayıtlar başarıyla tamamlandı. 'vki_kayitlari.txt' dosyasına kaydedildi.")

if __name__ == "__main__":
    main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime
import json
import os

MENU = {
    "meyveler": {
        "muz": 5,
        "çilek": 7,
        "ananas": 6,
        "mango": 8,
        "avokado": 10
    },
    "sütler": {
        "inek sütü": 4,
        "badem sütü": 6,
        "soya sütü": 5,
        "hindistan cevizi sütü": 7
    },
    "ekstralar": {
        "bal": 3,
        "fıstık ezmesi": 5,
        "chia tohumu": 4,
        "protein tozu": 8
    }
}

LOG_FILE = "smoothie_logs.json"

def menu_goster():
    print("\n--- YETBAR SMOOTHIE MENÜSÜ ---")
    for kategori, urunler in MENU.items():
        print(f"\n{kategori.upper()}:")
        for urun, fiyat in urunler.items():
            print(f"  - {urun.capitalize()}: {fiyat} TL")

def smoothie_olustur():
    secimler = {}
    toplam_fiyat = 0
    
    print("\nSmoothie'nizi oluşturalım!")
    
    for kategori in MENU.keys():
        print(f"\n{kategori.capitalize()} seçin:")
        menu_goster()
        
        while True:
            secim = input(f"Bir {kategori[:-1]} seçin (çıkmak için 'tamam' yazın): ").lower()
            if secim == "tamam":
                break
            if secim in MENU[kategori]:
                secimler.setdefault(kategori, []).append(secim)
                toplam_fiyat += MENU[kategori][secim]
                print(f"{secim.capitalize()} eklendi. Ara toplam: {toplam_fiyat} TL")
            else:
                print("Geçersiz seçim. Lütfen menüdeki ürünlerden birini girin.")
    
    return secimler, toplam_fiyat

def log_kaydet(secimler, toplam_fiyat):
    kayit = {
        "tarih": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "malzemeler": secimler,
        "fiyat": toplam_fiyat
    }
    
    if not os.path.exists(LOG_FILE):
        with open(LOG_FILE, "w") as f:
            json.dump([], f)
    
    with open(LOG_FILE, "r+") as f:
        logs = json.load(f)
        logs.append(kayit)
        f.seek(0)
        json.dump(logs, f, indent=4)

def satin_al():
    secimler, toplam_fiyat = smoothie_olustur()
    
    if not secimler:
        print("Hiçbir malzeme seçmediniz. İşlem iptal edildi.")
        return
    
    print("\nSeçimleriniz:")
    for kategori, urunler in secimler.items():
        print(f"{kategori.capitalize()}: {', '.join(urunler)}")
    
    print(f"\nToplam Fiyat: {toplam_fiyat} TL")
    
    onay = input("\nSatın almayı onaylıyor musunuz? (evet/hayır): ").lower()
    if onay == "evet":
        log_kaydet(secimler, toplam_fiyat)
        print("\nTeşekkür ederiz! Smoothie'niz hazırlanıyor. Afiyet olsun!")
        print(f"Toplam {toplam_fiyat} TL tutarındaki siparişiniz kaydedildi.")
    else:
        print("İşlem iptal edildi.")

def loglari_goster():
    if not os.path.exists(LOG_FILE):
        print("Henüz kayıtlı sipariş bulunmamaktadır.")
        return
    
    with open(LOG_FILE, "r") as f:
        logs = json.load(f)
    
    if not logs:
        print("Henüz kayıtlı sipariş bulunmamaktadır.")
        return
    
    print("\n--- GEÇMİŞ SİPARİŞLER ---")
    for i, log in enumerate(logs, 1):
        print(f"\nSipariş #{i} - {log['tarih']}")
        for kategori, urunler in log['malzemeler'].items():
            print(f"{kategori.capitalize()}: {', '.join(urunler)}")
        print(f"Toplam Fiyat: {log['fiyat']} TL")

def main():
    
    while True:
        print("\nANA MENÜ")
        print("1. Menüyü Görüntüle")
        print("2. Smoothie Satın Al")
        print("3. Geçmiş Siparişleri Görüntüle")
        print("4. Çıkış")
        
        secim = input("Lütfen bir işlem seçin (1-4): ")
        
        if secim == "1":
            menu_goster()
        elif secim == "2":
            satin_al()
        elif secim == "3":
            loglari_goster()
        elif secim == "4":
            print("YetBar'dan çıkılıyor. İyi günler!")
            break
        else:
            print("Geçersiz seçim. Lütfen 1-4 arasında bir numara girin.")

if __name__ == "__main__":
    main()
```
