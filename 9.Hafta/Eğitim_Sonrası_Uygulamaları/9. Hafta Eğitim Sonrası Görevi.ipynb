# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_acik_mi = False
        self.ses = 10
        self.kanal_listesi = ["NOW"]
        self.aktif_kanal = "NOW"

    def tv_ac(self):
        if not self.tv_acik_mi:
            self.tv_acik_mi = True
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_acik_mi:
            self.tv_acik_mi = False
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayar(self):
        while True:
            islem = input("Sesi artır: '>' | Sesi azalt: '<' | Çıkış: 'ç' >>> ")
            if islem == ">":
                if self.ses < 100:
                    self.ses += 1
                    print("Ses:", self.ses)
                else:
                    print("Ses zaten maksimum seviyede.")
            elif islem == "<":
                if self.ses > 0:
                    self.ses -= 1
                    print("Ses:", self.ses)
                else:
                    print("Ses zaten minimum seviyede.")
            elif islem == "ç":
                break
            else:
                print("Geçersiz işlem.")

    def kanal_ekle(self, yeni_kanal):
        if yeni_kanal not in self.kanal_listesi:
            self.kanal_listesi.append(yeni_kanal)
            print(f"{yeni_kanal} kanalı eklendi.")
        else:
            print(f"{yeni_kanal} zaten kanal listesinde.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal(self):
        self.aktif_kanal = random.choice(self.kanal_listesi)
        print(f"Şu anki kanal: {self.aktif_kanal}")

    def tv_bilgileri(self):
        print("\nTelevizyon Bilgileri:")
        print(f"Durum: {'Açık' if self.tv_acik_mi else 'Kapalı'}")
        print(f"Ses: {self.ses}")
        print(f"Kanal Listesi: {self.kanal_listesi}")
        print(f"Aktif Kanal: {self.aktif_kanal}\n")


kumanda = Kumanda()

while True:
    print("""
İşlemler:
1 - Televizyonu Aç
2 - Televizyonu Kapat
3 - Ses Ayarları
4 - Kanal Ekle
5 - Kanal Sayısını Göster
6 - Rastgele Kanala Geç
7 - Televizyon Bilgileri
q - Çıkış
""")

    secim = input("İşlem seçiniz: ")

    if secim == "1":
        kumanda.tv_ac()
    elif secim == "2":
        kumanda.tv_kapat()
    elif secim == "3":
        kumanda.ses_ayar()
    elif secim == "4":
        kanal_adi = input("Eklemek istediğiniz kanalın adını girin: ")
        kumanda.kanal_ekle(kanal_adi)
    elif secim == "5":
        kumanda.kanal_sayisi()
    elif secim == "6":
        kumanda.rastgele_kanal()
    elif secim == "7":
        kumanda.tv_bilgileri()
    elif secim.lower() == "q":
        print("Kumanda kapatılıyor...")
        break
    else:
        print("Geçersiz seçim.")
```
